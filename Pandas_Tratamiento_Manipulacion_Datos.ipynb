#**1. Entendiendo el Problema**


---




```python
import pandas as pd
```

```python
#Leer los datos de un archivo
#importando  datos

datos = pd.read_json('/content/datos_hosting.json')
datos.head()
```

```python
#Extraer los datos utilizando normalize para normalizar los datos json
#los va a llevar a un data frame
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

Vemosque nuestra tabla tine lista de elementos de numeros y de strings

#**2. Datos numéricos**


---




```python
datos.info()
```

```python
#Guardar en una  lista de las columnas
columnas = list(datos.columns)
columnas
```

```python
#Voy a explotarlas queremos que sean a partir de la posicion 3 en adelante
datos = datos.explode(columnas[3:])
datos
```

```python
#Si quiero que sea mi nuevo data frame no dee tener mi index ent lo vamos a resetear
#con rest_index, drop=true es para borrarlo de la memoria
datos.reset_index(inplace = True, drop = True)
datos.head()
```

```python
datos.info()
```

Cuando aparece object python no supo identificar si numero o texto
Por lo que todaia se debe mejorar esto

```python
import numpy as np
```

```python
#Para cambiar el tipo de datos, numericos (en este caso numeros enteros)
#Nos ayudamos de nampy. Transformar los datos y guardalos
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
#No va funcionar si hay algun registro con un valor que no sea entero y no va a realizar
```

```python
#Vemos que ya cambio a numeros enteros a traves del metodo anterior
datos.info()
```

```python
#Columnas que deso cambiar a enteras
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
#Aplicamos el metodo anterior para que sean enteros
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
#Ver si cambio el tipo de datos
datos.info()
```

```python
#Vemos que hay columanas numericas decimales
#Los gardamos  hacemos la transformacion
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
#Vemoss si se cambiaron los datos
#En el data frame
datos.info()
```

```python
#Columna  precio tiene un caracter por lo que se debe realizar lo siguiente
#Remover el dolar
#Examiar los valores de precio y con apply (aplicar una transformacion) usando la funcipo lamda
# landa recorera cada uno de los valores y lo va a reemplazar $ por '' nada
#llamar otro punto replace replazando si encuentro una como por nada
#con otra transformacion strip() quitar los espacioos que encuentres reemplazalos por nada
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
#Antes ya removimos los datos y ahora si aplicamos los cambios a los datos
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
#Ver que la columna precio ya se reemplazo
datos.info()
```

```python
#Quitar los caracters y comoas a las siguientes columnas
#Como vamos a examinar mas de 1(2,3,4,5) se debe aplicar doble corchete
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
#Aplicar la tranformacion a mi data frame
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
#Ver si hay cambios en mi data frame
datos.info()
```

#**3. Datos de texto**


---




```python
#Podemos extarer algunas palabras dentro de la tabla para poder realizar un análisis
datos.head()
```

```python
#tockenisacion es la division de texto en tokens e.i en partes mas pequeñas
#Transformar texto en palabras de la columna 'decripcion_local
# 'str' le decimos que es textor  con lower() devuelve una copia en minuscula
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
#Ver si ya se cambio en nestra tabla
datos.head()
```

```python
#Ver 'descripcion_local' examinar una variable aqui el 3169

datos['descripcion_local'][3169]
```

```python
#replace para reemplazar na cosas  adicionar parametro regex=true  para decirle que reemplace todo los caracteres especioles
#usando regex el regex [^a-zA-Z0-9\-\'] siempre va dentro de lo corchetes le estoy diciendo que reemplace todo
#menos las letras  de a a z y no sustituya los numeros , tambien no reemplacel el guion, tampoco las comillas
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
#Para cuando hay guines en medio de dos palabras
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
```

```python
datos['descripcion_local'].head()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
#Crear una lista
datos['comodidades'] = datos['comodidades'].str.split(',')
datos.head()
```

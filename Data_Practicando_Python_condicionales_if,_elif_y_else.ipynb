# Practicando Python: condicionales if, elif y else






## Condicionales if, else y elif

En Python, al igual que en otros lenguajes de programación, también contamos con estructuras condicionales. Esto significa que un bloque de código solo se ejecuta cuando se cumple una condición, es decir, cuando es verdadera. La estructura condicional más básica es el if. Vamos a utilizar Visual Studio Code para observar cómo sería la estructura de un if.

Lo primero que debemos escribir es if, que significa "si". Si algo ocurre, entonces...la condición que debe cumplirse o no.
Se crea automáticamente un espacio, que es la indentación. Esto indica que el código que sigue pertenece al if. Aquí colocamos la acción que se ejecutará si la condición establecida por el if se cumple.
¿Y si la condición no se cumple? Para eso, tenemos otra estructura que colocamos debajo de esa acción, que es el else.
Luego, colocamos una acción.
Es importante mencionar que el else es opcional; podemos incluirlo o no.


```python
"""
if condicion:
    accion
else:
    accion
"""
```

```python
nombre = 'Alura'
```

```python
if nombre == 'Alu':
    print('La condicion es verdadera')
else:
    print('La condicio es falsa')
```

Además de la condición del if, podemos querer otra condición más. Para eso, utilizamos la estructura elif, que colocamos entre el if y el else.


```python
if nombre == 'Alu':
    print('La condicion es verdadera')
elif nombre == 'Alura':
    print('La condicion es verdadera 2')
else:
    print('La condicio es falsa')
```

El programa debe recibir el peso y la altura de una persona y mostrar el valor del IMC, además de indicar si está por debajo del peso, con peso normal o por encima del peso. Crea un programa que reciba el peso (en kg) y la altura (en metros) y calcule el IMC usando la fórmula: IMC = peso / (altura ** 2)Luego, muestra el valor del IMC y un mensaje indicando si está por debajo del peso (IMC < 18.5), peso normal (18.5 <= IMC < 25) o por encima del peso (IMC >= 25).

Salida esperada:

image.png

```python
peso = int(input('Digite su peso (kg): '))
altura = float(input('Digite su altura (m): '))

imc =  peso / (altura ** 2)

print(f'Su IMC es: {imc}')

if imc <18.5 :
  print('Está por debajo del peso.')
elif  imc < 25:
  print('Tines un peso normal.')
elif imc >=25:
  print('Está por encima del peso.')
else:
  print('')
```

## Juego de múltiplos mágicos

Estás desarrollando un pequeño juego. El usuario ingresa un número entero y el programa debe evaluar lo siguiente:



*   Si el número es divisible por 3 y 5, muestra: `"¡Número mágico!"`
*   Si solo es divisible por 3, muestra: `"Divisible por 3"`
*   Si solo es divisible por 5, muestra: `"Divisible por 5"`
*   Si no es divisible por ninguno, muestra: `"No es un número mágico"`


Este tipo de lógica es muy útil en juegos, validaciones o filtros.

```python
entero = int(input('Ingresa un número entero: '))

if entero % 3 == 0 and entero % 5 == 0 :
  print('¡Número mágico!')
elif entero % 3 :
  print('Divisible por 3')
elif entero % 5 :
  print('Divisible por 5')
else:
  print('No es un número mágico')
```

## Sistema de clasificación para becas escolares

Una escuela otorga becas según tres criterios:

Ingreso familiar mensual.
Promedio del estudiante.
Asistencia (en porcentaje).
Reglas:

Si el ingreso es menor a \$1,500 y el promedio es mayor a 8.0 y la asistencia es al menos 90\% → "Beca completa"
Si el ingreso es menor a \$2,500 y promedio mayor a 7.0 y asistencia al menos 85\% → "Media beca"
En otros casos → "No elegible para beca"


```python
ingreso = float(input('Dígite su ingreso familiar mensual $:'))
promedio = float(input('Dígiite su promedio: '))
asistencia = float(input('Dígite su asistencia (porcentaje): '))

if ingreso < 1500 and  promedio > 8.0 and asistencia >= 90:
  print('Beca completa')
elif ingreso < 2500 and  promedio > 7.0 and asistencia >= 85 :
  print('Media beca')
else:
  print('No elegible para la beca')
```

## Sistema de evaluación de desempeño empresarial

Una empresa evalúa su trimestre con base en:
- Ingresos totales
- Gastos totales
- Número de nuevos clientes

Clasificación:


*   Si ingresos - gastos > \$10,000 y más de 50 nuevos clientes → "Trimestre Excelente"
*   Si ingresos - gastos > \$5,000 y al menos 20 clientes → "Trimestre Bueno"
*   Si ingresos - gastos > 0 → "Trimestre Regular"
*   Si ingresos - gastos ≤ 0 → "Trimestre Deficitario"

```python
ingresos_t = float(input('Dígite los ingresos totales $: '))
gastos_t = float(input('Dígite los gastos totales $: '))
num_de_nu_clientes = int(input('Dígite el número de nuevos clientes: '))

if (ingresos_t - gastos_t) > 10000 and num_de_nu_clientes > 50 :
  print('Trimestre Excelente.')
elif (ingresos_t - gastos_t) > 5000 and num_de_nu_clientes >= 20 :
  print('Trimestre Bueno.')
elif (ingresos_t - gastos_t) > 0 :
  print('Trimestre Regular.')
else:
  if(ingresos_t - gastos_t ) <= 0 :
    print('Trimestre Deficitario.')


```

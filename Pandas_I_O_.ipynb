# Aula 5 -Leyendo banco de datos

## Creando un banco de datos local

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect,text
#Generar el entorno
#Guarda la informacion los atributos tipos de datos..
#Crear tablas
#inspect para inspecionar que atributo qtengo dentro de las tablas
#text para convertir la querys en un tipo de dato en texto que se pueda entender dentro de la consulta

```

```python
#Crear el banco de datos local
#Utilizar uno de los bancos que ya viene instalado po default en google colab
#Existe otros mysql, oracle server...
#memory significa utilizar la memoria local de gooogle colab
engine = create_engine('sqlite:///:memory:')
```

```python
#engine es la base el motor de todo
```

## Escribiendo en un banco de datos


```python
import pandas as pd
```

```python
archivo = 'clientes_banco.csv'
```

```python
datos = pd.read_csv(archivo)
```

```python
datos.head(5)
```

```python
#Porque tenemos que sea csv para dataframe y ahora si a sql
datos.to_sql('clientes',engine,index=False)
```

```python
#Como puedo saber si funciono
#Revisa todo lo que hay dentro del motor
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

##Leyendo una consulta SQL

```python
#Cuando queremos todas las columnas ponemos select *
#from lo que quiero
query = 'SELECT * FROM clientes WHERE Categoria_de_renta = "Empleado"'
```

```python
#Consulta sql
#text se debe poner ya que debe ser un tipo de texto especial
#le digo que se conecte con conect
empleados = pd.read_sql(sql=text(query), con= engine.connect())
```

```python
type(empleados)
```

```python
empleados
```

```python
#Ya realizamos un filtro
#podemos guardarla la tabla

empleados.to_sql('empleados',con=engine.connect(),index=False, if_exists='replace')
```

```python
#Ya realizamos un filtro
#podemos guardarla la tabla

#empleados.to_sql('empleados',con=engine.connect(),index=False)
```

```python
pd.read_sql('empleados',con=engine.connect(),columns=['ID_Cliente','Grado_estudio','Rendimiento_anual'])
```

## Actulizando un banco de datos

```python
query = 'SELECT*FROM clientes'
```

```python
pd.read_sql(sql=text(query),con=engine.connect())
```

```python
#Borrado de registros
#identificar el tipo de error importamos una bibliote
from sqlalchemy.exc import SQLAlchemyError
query ='DELETE FROM clientes WHERE ID_Cliente=5008804'
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('#Registros borrados: ', r_set.rowcount)
```

```python
pd.read_sql('clientes', con=engine.connect())
```

```python
#quiero hacer una actualizacion
#5008808 ese cliente ya es nivel intermedio
#Actualizando registro2
query ='UPDATE clientes SET Grado_estudio="Nivel Superior" WHERE ID_Cliente=5008808'
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('#Registros actualizados: ', r_set.rowcount)

```

```python
pd.read_sql('clientes', con=engine.connect())
```
